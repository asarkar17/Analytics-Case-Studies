```python
#In this simple code, I extract the text from Jon Snow's wikipage. I changed "Kit Harington" to my name
#I did a bit of text preprocessing(stop words, lower case, puctuations, etc.)
#I did not do stemming in this one, you can add your own
#and output sorted word frequency and biagram frequency to two files
# I included biagrams so that I can catch phrases like "night watch", "night king", etc.

import bs4
from bs4 import BeautifulSoup # for grabing text from webpages
from nltk import bigrams
from urllib.request import urlopen #for webpages
import nltk
from string import punctuation # for removing punctuations
```

```python
def CleanText(wordMaps, string): #Here I will replace words I need to 
    for check in wordMaps:
        string=string.replace(check[0],check[1])
    return string

def WordFrequency(myOutFile,myText): # Here I count word frequencies(sorted) and output it to a table
    fdist = nltk.FreqDist(myText)
    sorted_fdist = sorted(fdist.items(), key=lambda x: x[1], reverse=True)
    myOutFile = open(myOutFile, 'w',encoding='utf-8')
    for f in sorted_fdist:
        print(f[0], f[1])
        myOutFile.write(f'{f[0]},{f[1]}')
        myOutFile.write('\n')
    myOutFile.close()  
    
def RemoveStopWords(stopwords, myText): # Here I remove all the stopwords
    NewText = [w for w in myText if w not in stopwords]
    return NewText    

def Bigram (tokens): # I make Bigrams here so I can catch stuff like "Knight King"
    bigram=bigrams(tokens)
    return bigram
```

```python
url = "https://en.wikipedia.org/wiki/Jon_Snow_(character)"
html = urlopen(url).read()
#print (html) 
soup = BeautifulSoup(html, "html.parser")

wordMaps=[("\'s",""),("Kit Harington","TIANJIE DENG")]# I want to change the actor's name into my name, add more word pairs to this list if you want
soup = BeautifulSoup(html, "html.parser")
raw=[]
body = soup.find("div", {"id":"bodyContent","class":"mw-body-content"})
para = body.findAll("p")

for p in para:
    ptext=p.get_text().strip()
    newText=CleanText(wordMaps, ptext) # I changed the actor's name into my name, as well as deleting "\'s'"
    raw.append(newText)
raw="\n".join(raw).lower()
tokens=nltk.word_tokenize(raw)
stopwords = nltk.corpus.stopwords.words('english')
#print(stopwords)
#tokens=RemoveStopWords(stopwords, tokens)
#WordFrequency(tokens) # I ran the frequency first, to see the top words, in order to find more stop words.
custStopwords=set(["the", "of"]) # based on the above frequency, I decided to add more stop words
punct_set = set(list(punctuation)) 
stops = set(stopwords).union(custStopwords).union(punct_set) # this set include all english stop words, my newly added stop words, as well as punctuations

tokens=RemoveStopWords(list(stops), tokens)
mywordOutFile="C:\\Users\\Tianjie.Deng\\Dropbox\\Work\\Teach\\INFO4360_Spring 2019\\Python\\Topic 2 Text Processing\\Lecture Notes\\wikiFreq.txt"
WordFrequency(myOutFile,tokens)

bigram=Bigram(tokens)
#print(list(bigram))
mybigramOutFile="C:\\Users\\Tianjie.Deng\\Dropbox\\Work\\Teach\\INFO4360_Spring 2019\\Python\\Topic 2 Text Processing\\Lecture Notes\\wikiBigramFreq.txt"
#WordFrequency(myOutFile,tokens)
WordFrequency(mybigramOutFile,bigram)
```
